```python
test_data = data
design_facts = []
approaches = [RA_t1_out,RA_t4_out,RA_t3_out,RA_t2_out]
#approaches = [optimal_insample,lt_second_stage_sols]
approach_labels = ['RA-Customer','RA-Lvl.1','RA-Lvl.2','RA-Lvl.3']

for idx_app in range(len(approaches)):
    approach = approaches[idx_app]
    #Counting the number of arcs used in total
    approach_facts = {}
    approach_facts['Total arcs'] = sum(approach['y'].values())
    
    sum_flows = 0
    em_flows = 0
    planned_capacity = 0

    entity_count = 0
    entity_caps = 0

    averages = 0
    counts = 0
    entities = []
    for i in test_data['V']:
        if i not in test_data['v_subsets']['Retail']:
            all_flows = 0
            for j in test_data['downstream'][i]:
                for k in test_data['K']:
                    if (i,j,k) in test_data['valid_arcs']:
                        all_flows += approach['x'][i,j,k]
            
            if all_flows > 0:
                entity_count += 1

            entity_caps += all_flows/test_data['q_mix'][i]

    
    for (i,j,k),y in approach['y'].items():
        if y == 1:
            entities.append(i)
            entities.append(j)
            sum_flows += approach['x'][i,j,k]
            planned_capacity += approach['x'][i,j,k]/test_data['q_ind'][i,j,k]
            #print(planned_capacity)
            inner_averages = 0
            for xi in test_data['Omega']:
                em_flows += approach['z'][i,j,k,xi]
                
                if approach['z'][i,j,k,xi] > 0:
                    inner_averages+=1/100
                    averages += inner_averages
            if inner_averages > 0:
                counts += 1
            
        #print(planned_capacity)
                
        #print(inner_averages)
    entities = list(set(entities))

    approach_facts['Total entities used'] = int(len(entities))
    approach_facts['Total planned flow'] = int(sum_flows)
    approach_facts['Average emergent flows'] = int(em_flows/len(test_data['Omega']))
    approach_facts['Average arc utilization'] = np.round(planned_capacity/approach_facts['Total arcs'],2)
    approach_facts['Average entity utilization'] = entity_caps/entity_count
    approach_facts['% of scenarios with emergent flows'] = averages/counts
    design_facts.append(approach_facts)
pd.DataFrame(design_facts,index=approach_labels).T
```
